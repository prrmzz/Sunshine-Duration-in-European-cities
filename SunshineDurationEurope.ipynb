```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('sunshine.xlsx')
data.head()
```

```python
location = data['City']+', '+data['Country']
data_mounth = data.iloc[:,2:-1]
mounth_names = data_mounth.columns
data_mounth = np.array(data_mounth)
```

# This is a heatmap of the sunshine duration in europe :
* by looking to this plot, you can easily recognize which cities/countries has
longer sunshine than others in different mounths.

```python
plt.figure(figsize=(20, 80))
sns.heatmap(data_mounth, annot=True, cmap='viridis', cbar=False)
plt.title('Heatmap of sunshine duration in european cities by Month')
plt.xlabel('Month')
plt.ylabel('City')
plt.xticks(ticks=range(len(mounth_names)), labels=mounth_names)
plt.yticks(ticks=range(len(location)), labels=location)
plt.yticks(rotation=25)
plt.show()
```
